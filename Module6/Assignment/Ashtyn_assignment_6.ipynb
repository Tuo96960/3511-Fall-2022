Problem 8

Determine the age that each of the prize winners was when they won the prize. Find the youngest and oldest to win the prize. Find the mean age to win the prize, and the standard deviation of it. Make a histogram of the age each person was when they won the prize.

Hint: You can use Pandas DatetimeIndex method to extract the year from the date format yyyy-mm-dd

pd.DatetimeIndex(date).year

Where date is given in a formate like year-month-day, the above will returen the year, hence the .year in the call above. Use 'shit+tab' on pandas.DatetimeIndex to learn more.

```python
#Determine the age that each of the prize winners was when they won the prize
df.dropna()
bd_array = df[['birth_date']].to_numpy()
b = pd.DatetimeIndex(df.birth_date).year
age = df.year - b

#Find the youngest and oldest to win the prize
age1 = sorted(age)
#print(age)
#print(len(age))
print('\n The youngest age of a Nobel Award recipient is:', age[0])
print('\n The oldest age of a Nobel Award recipient was:', age[909], '\n')

#Find the mean age to win the prize
mean_age_array = []
for i in age1:
  if 25 <= i <= 90:
    mean_age_array.append(i)
  else:
    continue
mean = np.mean(mean_age_array)
print('The mean of the Recipient ages is:', mean)

#find st d
numerator_sum = 0
for i in mean_age_array:
  numerator_sum += (i - mean)**2

eq = np.sqrt(numerator_sum/len(mean_age_array))
print('\n The Standard Deviation is:', eq, '\n')

#Make a histogram of the age each person was when they won the prize
fig = plt.figure()
c = fig.add_axes([0.1,0.1,0.8,0.8])
c.set_title('Ages of Nobel Award Recipients')
c.set_xlabel('Age')
c.set_ylabel('Number of Recipients')
c.hist(age, bins=100, color='gray' ,ec='black')
```
