Define a function called MakeHist that will make a histogram and takes as its arguments a numpy array (data), integer (number of bins), and strings for its title and horizontal and vertical axis titles. Once you have your MakeHist function defined, create a numpy array generated from the random normal distribution centered at zero. **Call** your MakeHist function 3 times with bin values of 10, 100, and 1000. Each call should use your random numpy array as the input data. For this exersice you can select any name for the horizontal and vertical axes, but set the histogram title to the number of bins it was produced with.

```python
import matplotlib.pyplot as plt
import numpy as np

#call makehist 3 times w/ 10, 100, 1000

r = np.random.randn(1000)

def main():
  MakeHist(r=r, x=10, title = "nbins = 10", xaxis = "x", yaxis = "counts")
  MakeHist(r=r, x=100, title = "nbins = 100", xaxis = "x", yaxis = "counts")
  MakeHist(r=r, x=1000, title = "nbins = 1000", xaxis = "x", yaxis = "counts")

def MakeHist(r, x, title, xaxis, yaxis):
  fig = plt.figure()
  h = fig.add_axes([0.1,0.1,0.8,0.8])
  h.set_title(title)
  h.set_xlabel(xaxis)
  h.set_ylabel(yaxis)
  h.hist(r,bins=x)


main()
```
